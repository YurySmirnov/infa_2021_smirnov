## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
a, b = 1, 1
```

```python
def f(x):
  global a
  global b
  return (x**b + a**b) / x**b
```

```python
xnegat = np.setdiff1d(np.linspace(-100, 0, num=30000), [0])
xposit = np.setdiff1d(np.linspace(0, 100, num=30000), [0])
a, b = 1, 1
ynegat = f(xnegat)
yposit = f(xposit)
fig = plt.figure(figsize=(20,20))
axes = fig.add_subplot(111)


c = plt.plot(xnegat, ynegat, label='A = 1, B = 1')
c2 = plt.plot(xposit, yposit, c[0].get_color())

a, b = 2, 1
ynegat = f(xnegat)
yposit = f(xposit)
d = plt.plot(xnegat, ynegat, label='A = 2, B = 1')
d2 = plt.plot(xposit, yposit, d[0].get_color())

a, b = 1, 2
ynegat = f(xnegat)
yposit = f(xposit)
e = plt.plot(xnegat, ynegat, label='A = 1, B = 2')
e2 = plt.plot(xposit, yposit, e[0].get_color())
plt.legend()

axes.set_ylim((-50,80))
axes.set_xlim((-5,5))
```
